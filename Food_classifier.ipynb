```python
labelA = 'red_velvet_cake'
labelB = 'carrot_cake'
```

```python
for img in get_image_files(foodPath):
  if labelA in str(img):
    img.rename(f"{img.parent}/{labelA}_{img.name}") # food name / numbers
  elif labelB in str(img):
    img.rename(f"{img.parent}/{labelB}_{img.name}")
  elif 'carrot cake 251354' in str(img):
    os.remove(img)
  else:
    os.remove(img)
```

```python
def GetLabel(fileName):
  if len(fileName) == 22:
    return ' '.join(fileName.split('_')[:3])
  else:
    return ' '.join(fileName.split('_')[:2])

```

```python
dls = ImageDataLoaders.from_name_func(
    foodPath, get_image_files(foodPath), valid_pct = 0.2, seed = 420,
    label_func=GetLabel, item_tfms=Resize(224)
)
```

ResNet34(# of leayers)

```python
learn = cnn_learner(dls, resnet34, metrics = error_rate, pretrained=True)
learn.fine_tune(epochs = 10)  #. An epoch is one complete pass through the entire training dataset
```

```python
from google.colab import files
uploader = files.upload()

for img in uploader.items():
  uploadedImg = img[0]
img = PILImage.create(uploadedImg)
img.show()

label,_,probs = learn.predict(img)
if label == 'red velvet':
  print(f"This is a {label} cake.")
else:
  print(f"This is a {label}.")

print(f"{labelA} {probs[1].item():.6f}")
print(f"{labelB} {probs[0].item():.6f}")
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
interp.plot_top_losses(10)
```
